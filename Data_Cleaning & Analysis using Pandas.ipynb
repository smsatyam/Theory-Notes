# Data Cleaning

Now that we understand our data, we are in the second stage. Before analyzing data , we need to clean our data for the clean analysis.

# What we have now ?

We have missing values within the Income variable
There is also wrong formatting for Dt_Customer column
Duplicated row
Letâ€™s see how to tackle these issues with pandas.

# dropna()

df.dropna(inplace=True)
dropna() method removes all the rows within the dataset.

inplace argument gives us option whether we permanently drop the null values or not.

Note: Without inplace argument, we can also assign a variable to create new dataframe for example df1 = df.dropna()

```python
df.dropna(inplace=True)
```

```python
mean = df.Income.mean() # average of Income
df['Income'].fillna(mean,inplace=True)
```

# fillna()

fillna() method fills the null values desired value. For numerical columns, we can use mean or median, for categorical values we can use most frequent values (mode) to impute the values.

# drop_duplicates()

By default, drop_duplicates method removes duplicate rows based on all columns.

```python
df.drop_duplicates(inplace=True) # inplace = True means we removed permanently.
```

# to_datetime()

to_datetime() function converts an object to datetime format.

```python
df['Dt_Customer'] =  pd.to_datetime(df['Dt_Customer'])
```

```python
#If we want to create a new column named year, month or day of the week we can also use the following functions.

```

```python
df['year_Customer'] =  df['Dt_Customer'].dt.year
df['month_Customer'] = df['Dt_Customer'].dt.month ## month of the customers
df['day_Customer'] = df['Dt_Customer'].dt.dayofweek ## dayofweek as a number 0 = Monday  6 = Sunday
df['Dt_Customer'].dt.strftime('%A') # This shows full weekday name
```

# sort_values()

If we want to sort our customers by their income. We can use the sort_values method. This method allows us to sort our data frame by any column and any order ascending or descending.

```python
df.sort_values(by='Income', ascending=False)
```

# query()

query() method filters out the data frame by the condition we want.

Letâ€™s say we want to see only customers who have income higher than 100k.

```python
df.query('Income > 100000') 
```

```python
#Alternative way :

df[df['Income'] > 100000]
```

# groupby()

group by method aggregates our values by splitting the objects.

```python
df.groupby('Education')['Income'].mean()
```

```python
df.groupby('Education').agg({'Income' : 'mean'})
```

# pivot_table()

pivot_table() method creates a useful pivot table for us.

There are 4 arguments we should use as input

data, index, columns and values.

By default, the method uses the mean as an aggregation function. We can also change it.

```python
pd.pivot_table(data = df, values= 'Income', index = 'Education', columns = 'Marital_Status')
```

# apply()

apply method gives us the ability to use any function inside. We can easily make any calculation with the apply method.

Suppose We want to convert numbers to strings Accepted and Not Accepted from our Customers Response.

```python
df['Response'].apply(lambda x : 'Accepted' if x == 1 else 'Not Accepted' )
```

# replace()

```python
df['Marital_Status'].replace(to_replace=['Alone','Divorced','Widow','YOLO','Absurd'], value='Single')
```
